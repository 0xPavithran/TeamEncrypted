```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Number of points for each class
num_points = 100

# Generate random data
np.random.seed(42)
data = {'Random_X': np.random.rand(3 * num_points),
        'Random_Y': np.random.rand(3 * num_points),
        'Class': np.repeat(['a', 'b', 'c'], num_points)}

df = pd.DataFrame(data)

# Define different colors for each class
class_colors = {'a': 'red', 'b': 'green', 'c': 'blue'}

# Create a figure and axis
fig, ax = plt.subplots()

# Cluster and connect nodes with a single connection for each node
for class_label, color in class_colors.items():
    color_df = df[df['Class'] == class_label]
    
    # Connect each node to the next node in the same color
    for i in range(len(color_df) - 1):
        ax.plot([color_df['Random_X'].iloc[i], color_df['Random_X'].iloc[i+1]],
                [color_df['Random_Y'].iloc[i], color_df['Random_Y'].iloc[i+1]],
                color=color, linestyle='-', linewidth=1)

    # Scatter plot for nodes
    ax.scatter(color_df['Random_X'], color_df['Random_Y'], marker='o', s=50, c=color, alpha=0.7)

# Customize the plot
plt.title('')
plt.xlabel('')
plt.ylabel('')

# Remove x and y axis labels
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```
